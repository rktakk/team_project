### Population Growth Analysis

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load the image using a relative path
img = mpimg.imread('Graph-Analysis-Raj/Graphs-Colour.png')

# Display the image
plt.imshow(img)
plt.axis('off')  # Hide axes
plt.show()
```

The above graph has year on the x-axis and the selected variables on the y-axis. Each graph also shows the total population growth over time. 

##### Fertility Rate (Births per Woman)
For the first graph fertility rate total, even though the births per women are on a downward trend the total population growth indicates that the population is still growing. From this we can draw the conclusion that the population growth is not dependant on the fertility rate alone. 

##### Net Migration
The second graph net migration, from the late 1960s to around 1987 total population growth is steady despite migration being low. However, starting in 1987 net migration steadly increases as does the population. Between 2017 and 2018 there is a spike in net migration and around 2020 there is a crash in net migration which can most likely be attributed to Covid-19. From this graph we can conclude net migration does influence population growth. But to understand the full extent a more in depth analysis would need to be conducted. 

##### Population Growth (Annual %)
The third graph population growth shows that despite spikes and dips the annual population growth % continues to steadly increase. In 2020 the annual population growth % does seem to level off. However, again this can most likely be attribute to Covid-19. Data from additional years is required to understand if the population stopped growing or if it continues to grow. 

#### Closing Remarks
In reviewing the above graphs and the analysis above, population growth does not rely on a single variable. Rather it relies on multiple. In order to fully understand the trend of population growth in Canada I believe more variables such as economic and large scale healthcare events should be considered.
